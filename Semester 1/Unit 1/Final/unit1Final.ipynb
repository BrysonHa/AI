# What is the relationship between the cost of a book and its average rating on the Amazon Best Seller List?

To answer this question, I will conduct a retrospective observational study using census data on Amazonâ€™s best-selling books from 2010 to 2020. The dataset was compiled by scraping Amazon, with missing information supplemented using the Goodreads API. One possible source of bias is that some books appear multiple times because they were listed on more than one bestseller list, which could slightly skew the results.

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('books.csv')
```

```python
ax = sns.scatterplot(data = df, x = "Price", y = "User Rating")

r = df['Price'].corr(df['User Rating'])

ax.text(0.95, 0.15, "r: " + str(r),
        horizontalalignment='right',
        verticalalignment='top',
        transform=ax.transAxes,
        fontsize=12,
        color="red")

ax.text(0.95, 0.1, "r^2: " + str(r**2),
        horizontalalignment='right',
        verticalalignment='top',
        transform=ax.transAxes,
        fontsize=12,
        color="red")
```

There is almost no relationship between the price of books and their user rating on the Amazon Best Seller list. This is supported by the small negative Pearson correlation coefficient of â€“0.133, which indicates a very weak negative linear relationship between cost and average rating. The coefficient of determination further supports this conclusion, showing that only about 1.771% of the variation in ratings can be explained by price, a negligible relationship. The scatter plot also shows no clear relationship.
