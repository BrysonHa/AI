What is the relationship between the number of Yelp reviews of a Denver ramen restaurant compared to its number of Google reviews?

To answer this question I will use data provided by my teacher, which she took directly from yelp and google. This will be a retrospective observational study. Some potential biases may be that the data set may not be entierly complete, or that some of the data may be innacurate due to human error.

```python
import pandas
import seaborn
```

```python
df = pandas.read_csv("denverRamen.csv")
df = df.drop(1) #Drop Outlier

df["Yelp Reviews"] = pandas.to_numeric(df["Yelp Reviews"], errors="coerce")
df["Google Reviews"] = pandas.to_numeric(df["Google Reviews"], errors="coerce")
```

Removed outlier Jinya Ramen, google ratings were way above the rest

```python
ax = seaborn.scatterplot(data = df, x = "Yelp Reviews", y = "Google Reviews")
seaborn.regplot(data = df, x = "Yelp Reviews", y = "Google Reviews", ci = None, line_kws = {"color": "red"})

r = df["Yelp Reviews"].corr(df["Google Reviews"])
r2 = r ** 2

ax.text(0.95, 0.95, "r: " + str(r),
        horizontalalignment='right',
        verticalalignment='top',
        transform=ax.transAxes,
        fontsize=12,
        color="red")

ax.text(0.95, 0.90, "r^2: " + str(r2),
        horizontalalignment='right',
        verticalalignment='top',
        transform=ax.transAxes,
        fontsize=12,
        color="red")
```

The correlation between google and yelp reviews of ramen resturants in denver is a high liner positive correlation. This is because the pearson correlation constant is roughly, 0.902 where a value of 1 is perfectly postively linear, so the correlation is highly linear and positive. The line of best fit aligns with 81.278% of the data, due to the coefficient of determination, concluding that it is a pretty good representation of the data.
